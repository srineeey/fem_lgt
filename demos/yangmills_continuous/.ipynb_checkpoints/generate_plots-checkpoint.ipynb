```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np
import os
import glob
import json
```

```python
def parse_folder(folder_name):
    print(folder_name)
    folder_dict = {}
    folder_dict["dir"] = folder_name
    
    #folder_pars = folder_name.split("_")
    folder_pars = (folder_name.split("/")[-1]).split("_")
    
    folder_dict["pars"] = {}
    
    #print(f"{folder_name}/pars.json")
    #with open(f"{folder_name}/pars.json", "r") as f:
    #    folder_dict["pars"] = json.load(f)
    
    
    for i, par_name in enumerate(par_names.keys()):
        folder_dict["pars"][par_name] = par_names[par_name][1](folder_pars[i].replace(par_name,""))
       
    folder_dict["pars"]["h"] = 1./folder_dict["pars"]["n"]
    
    print(folder_dict["pars"])
    
    folder_dict["obs"] = {}
    for obs_name in obs_names.keys():
        
        files = glob.glob(f"{folder_name}/{obs_name}_*.json")
        print(files)
        
        if len(files) != 1:
            raise ValueError(f"observable name '{obs_name}' has no or multiple matching files:{files}")
        
        with open(files[0],"r") as obs_f:
            #obs = json.load(obs_f)
            obs = np.array(json.load(obs_f))
        
        folder_dict["obs"][obs_name] = obs
        
    
    return folder_dict
```

```python
global_folder_name = "results/"
folder_names = os.listdir(global_folder_name)

folder_names = [ global_folder_name + folder_name for folder_name in folder_names]

folder_dicts = []
par_names = {"n":("n", int), "order":("n", int), "dt":("$\tau$", float)}
#obs_names = {"C":r"$\rho$", "dq2":r"$\Delta\rho$", "H":r"$\mathcal{H}$"}
obs_names = {"C":r"$\rho$", "dq2":r"$\Delta\rho$", "H":r"H", "HBL":r"H_{\mathcal{B},L}", "HBT":r"H_{\mathcal{B},T}", "HEL":r"H_{\mathcal{E},L}", "HET":r"H_{\mathcal{E},T}"}


for folder_name in folder_names:
    folder_dicts.append(parse_folder(folder_name))

```
